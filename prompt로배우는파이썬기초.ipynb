#### 2. 변수
변수는 값을 저장하는 데 사용되는 이름이며, 데이터를 참조하고 조작하는 데에 필요한 기본적인 요소입니다. 파이썬에서 변수를 사용하여 데이터를 저장하고, 이를 활용하여 계산하거나 조작할 수 있습니다. 변수는 다양한 데이터 타입을 가질 수 있으며, 값이 변경될 수 있는 특징을 가지고 있습니다.

변수를 사용하기 위해서는 변수를 선언하고 값을 할당해야 합니다. 파이썬에서는 변수 선언을 별도로 할 필요가 없으며, 변수에 값을 할당하는 과정에서 변수가 자동으로 생성됩니다. 변수 할당은 등호(=)를 사용하여 수행되며, 오른쪽의 값을 왼쪽의 변수에 저장합니다.

다음은 변수를 사용한 예제입니다.



```python
# 변수 할당
name = "Alice"
age = 25
pi = 3.14
is_student = True

# 변수 값 출력
print(name)         # "Alice" 출력
print(age)          # 25 출력
print(pi)           # 3.14 출력
print(is_student)   # True 출력

# 변수 값 변경
name = "Bob"
age = age + 1
pi = pi * 2
is_student = False

# 변경된 변수 값 출력
print(name)         # "Bob" 출력
print(age)          # 26 출력
print(pi)           # 6.28 출력
print(is_student)   # False 출력

```

위의 예제에서는 name, age, pi, is_student라는 변수를 선언하고 값을 할당했습니다. 각 변수에는 문자열, 정수, 부동 소수점 숫자, 불리언(Boolean) 값이 할당되었습니다. 변수 값은 print() 함수를 사용하여 출력되었습니다.

#### 4. 연산자
파이썬에서 연산자는 값을 조작하거나 연산을 수행하는 데 사용되는 기호입니다. 파이썬은 다양한 종류의 연산자를 제공하여 수치 연산, 논리 연산, 비교 연산 등을 수행할 수 있습니다.


1. 산술 연산자:
- `+` : 덧셈 연산을 수행합니다.
- `-` : 뺄셈 연산을 수행합니다.
- `*` : 곱셈 연산을 수행합니다.
- `/` : 나눗셈 연산을 수행하고 결과를 실수형으로 반환합니다.
- `//` : 나눗셈 연산을 수행하고 결과를 정수형으로 반환합니다.
- `%` : 나머지 연산을 수행합니다.
- `**` : 거듭제곱 연산을 수행합니다.

```python

a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.3333333333333335
print(a // b) # 3
print(a % b)  # 1
print(a ** b) # 1000
```


1. 할당 연산자:
- `=` : 변수에 값을 할당합니다.
- `+=` : 현재 변수에 값을 더하여 결과를 변수에 할당합니다.
- `-=` : 현재 변수에서 값을 빼고 결과를 변수에 할당합니다.
- `*=` : 현재 변수에 값을 곱하여 결과를 변수에 할당합니다.
- `/=` : 현재 변수를 나누고 결과를 변수에 할당합니다.
- `//=` : 현재 변수를 나눈 뒤 정수형으로 변환하여 결과를 변수에 할당합니다.
- `%=` : 현재 변수를 나눈 뒤 나머지를 결과로 변수에 할당합니다.

```python

x = 10

x += 5  # x = x + 5
print(x)  # 15

x -= 3  # x = x - 3
print(x)  # 12

x *= 2  # x = x * 2
print(x)  # 24

x /= 4  # x = x / 4
print(x)  # 6.0

x //= 2 # x = x // 2
print(x)  # 3

x %= 2  # x = x % 2
print(x)  # 1
```


1. 비교 연산자:
- `==` : 두 값이 같은지 비교합니다.
- `!=` : 두 값이 다른지 비교합니다.
- `>` : 왼쪽 값이 오른쪽 값보다 큰지 비교합니다.
- `<` : 왼쪽 값이 오른쪽 값보다 작은지 비교합니다.
- `>=` : 왼쪽 값이 오른쪽 값보다 크거나 같은지 비교합니다.
- `<=` : 왼쪽 값이 오른쪽 값보다 작거나 같은지 비교합니다.

```python

a = 5
b = 3

print(a == b)  # False
print(a != b)  # True
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False
```


1. 논리 연산자:
- `and` : 두 조건식이 모두 참일 때 참을 반환합니다.
- `or` : 두 조건식 중 하나라도 참이면 참을 반환합니다.
- `not` : 조건식의 결과를 반대로 반환합니다.

```python

x = 5
y = 3
z = 7

print(x > y and x < z)  # True
print(x > y or x > z)   # True
print(not(x > y))       # False
```



위의 예제에서는 주요한 연산자들을 다루었습니다. 파이썬은 다양한 연산자를 제공하므로 필요에 따라 추가적인 연산자를 학습하고 활용할 수 있습니다.

````python
a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.3333333333333335

x = 10

x += 5  # x = x + 5
print(x)  # 15

x -= 3  # x = x - 3
print(x)  # 12

a = 5
b = 3

print(a == b)  # False
print(a != b)  # True
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False
````

#### 5. 조건문

파이썬에서 조건문은 주어진 조건에 따라 코드 블록을 실행하거나 건너뛰는 데 사용됩니다. 주로 `if`, `elif`, `else` 키워드를 사용하여 작성합니다. 조건문은 프로그램이 특정 조건을 충족할 때 특정 동작을 수행하도록 하는데 유용합니다. 아래 예시를 통해 조건문을 자세히 살펴보겠습니다.



```python

age = 18

if age >= 20:
    print("성인입니다.")
elif age >= 13:
    print("청소년입니다.")
else:
    print("어린이입니다.")
```



위의 예제에서는 `age` 변수에 값을 할당하고, 해당 값에 따라 조건문이 실행됩니다.
- 첫 번째 조건 `age >= 20`은 참이 아니므로 건너뛰고 다음 조건으로 넘어갑니다.
- 두 번째 조건 `age >= 13`은 참이므로 해당 코드 블록이 실행되고 "청소년입니다."라는 메시지가 출력됩니다.
- 마지막으로 `else` 구문은 앞선 조건들이 모두 거짓일 경우 실행되며, 위 예제에서는 해당되지 않으므로 실행되지 않습니다.

조건문은 단일 `if` 문으로 시작할 수도 있고, 여러 개의 `elif` 문과 선택적인 `else` 문을 포함할 수도 있습니다. 이를 통해 여러 가지 조건을 처리하고 원하는 결과를 얻을 수 있습니다.

아래 예시는 조건문을 사용하여 성적에 따라 학점을 부여하는 예제입니다.

```python

score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print("학점:", grade)
```



위 예제에서는 `score` 변수에 값을 할당하고, 해당 값에 따라 조건문이 실행됩니다. `if`와 `elif` 조건들은 참인 경우 해당되는 학점을 `grade` 변수에 할당하고, `else` 조건은 모든 이전 조건들이 거짓일 경우 실행되어 'F'를 `grade` 변수에 할당합니다. 마지막으로, 학점이 출력됩니다.

이러한 방식으로 조건문을 사용하여 프로그램의 흐름을 제어하고, 주어진 조건에 따라 다른 동작을 수행할 수 있습니다.

````python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print("학점:", grade)
````

while 조건식:

#### 7. 함수

함수는 다음과 같은 형식으로 정의됩니다:



```python

def 함수이름(매개변수):
    # 함수의 동작을 정의하는 코드
    # 필요에 따라 매개변수를 활용하고, 결과를 반환할 수 있음
    return 결과값
```



위의 형식에서 매개변수는 함수에 전달되는 입력 값이며, 함수 내부에서 사용될 수 있습니다. 결과값은 함수가 실행된 후 반환되는 값입니다.

다음은 간단한 함수의 예제입니다. 이 함수는 두 개의 숫자를 더한 결과를 반환합니다:

```python

def add_numbers(a, b):
    sum = a + b
    return sum
```



위의 예제에서 `add_numbers` 함수는 `a`와 `b`라는 두 개의 매개변수를 받고, 덧셈 연산을 수행한 뒤 그 결과를 `return` 키워드를 통해 반환합니다.

이제 이 함수를 호출하여 결과를 확인해봅시다:

```python

result = add_numbers(3, 5)
print(result)
```



출력 결과는 `8`이 될 것입니다. 함수를 호출할 때 인자로 `3`과 `5`를 전달하면, 함수 내부에서는 `a`에 `3`, `b`에 `5`가 할당되어 덧셈 연산이 수행되고, 결과로 `8`이 반환됩니다.

이처럼 함수는 입력을 받아서 어떠한 작업을 수행하고 결과를 반환하는 구조로 사용됩니다. 실제로 함수는 더 복잡한 작업을 수행할 수 있으며, 필요에 따라 매개변수를 조정하거나 반환값을 다양하게 활용할 수 있습니다.

````python
def add_numbers(a, b):
    sum = a + b
    return sum

result = add_numbers(3, 5)
print(result)
````

#### 8. 리스트
파이썬에서 리스트(List)는 여러 개의 값을 저장하는 데이터 구조입니다. 리스트는 대괄호([])로 둘러싸여 있으며, 각 값은 쉼표로 구분하여 나열됩니다. 리스트는 순서가 있으며, 값의 추가, 삭제, 수정 등이 가능합니다. 아래는 리스트의 간단한 설명과 예제입니다.

리스트를 만들기 위해서는 대괄호([])를 사용하고, 리스트 내부의 값들은 쉼표(,)로 구분합니다. 예를 들어, 1부터 5까지의 숫자를 저장하는 리스트를 만들어보겠습니다.

```python

numbers = [1, 2, 3, 4, 5]
```



위의 코드에서 `numbers`라는 이름의 리스트를 생성하였습니다. 이 리스트에는 1, 2, 3, 4, 5라는 값이 순서대로 저장되어 있습니다.

리스트는 다양한 데이터 타입을 저장할 수 있습니다. 예를 들어, 숫자, 문자열, 불리언 값 등을 혼합하여 리스트에 저장할 수 있습니다.

```python

mixed_list = [1, "Hello", True, 3.14]
```



위의 예제에서 `mixed_list`라는 이름의 리스트를 생성하였습니다. 이 리스트에는 숫자 1, 문자열 "Hello", 불리언 값 True, 그리고 실수 3.14가 순서대로 저장되어 있습니다.

리스트에는 인덱스(Index)를 사용하여 개별 요소에 접근할 수 있습니다. 인덱스는 0부터 시작하며, 대괄호([]) 안에 인덱스 값을 넣어주면 해당 인덱스에 위치한 요소를 가져올 수 있습니다.

```python

numbers = [1, 2, 3, 4, 5]
print(numbers[0])  # 첫 번째 요소에 접근
print(numbers[2])  # 세 번째 요소에 접근
```



위의 코드는 `numbers` 리스트의 첫 번째 요소와 세 번째 요소를 출력하는 예제입니다. 출력 결과는 각각 1과 3이 됩니다.

리스트에는 다양한 메서드(method)가 제공되어 값의 추가, 삭제, 검색 등을 수행할 수 있습니다. 예를 들어, `append()` 메서드를 사용하여 리스트의 끝에 값 하나를 추가할 수 있습니다.

```python

fruits = ["apple", "banana", "cherry"]
fruits.append("orange")  # "orange"를 리스트의 끝에 추가
print(fruits)
```



위의 코드에서 `fruits` 리스트에 "orange"를 추가하는 예제입니다. `append()` 메서드를 사용하여 "orange"를 리스트의 끝에 추가하였습니다. 출력 결과는 `["apple", "banana", "cherry", "orange"]`가 됩니다.

리스트에 대한 다양한 작업과 메서드에 대해서는 파이썬의 공식 문서나 파이썬 학습 자료에서 자세히 확인할 수 있습니다.

````python
numbers = [1, 2, 3, 4, 5]
print(numbers[0])  # 첫 번째 요소에 접근
print(numbers[2])  # 세 번째 요소에 접근

fruits = ["apple", "banana", "cherry"]
fruits.append("orange")  # "orange"를 리스트의 끝에 추가
print(fruits)
````
