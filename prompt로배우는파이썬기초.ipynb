````python
def uppercase(func):
    def wrapper():
        result = func()  # ÏõêÎûò Ìï®Ïàò Ìò∏Ï∂ú
        return result.upper()  # Î∞òÌôòÍ∞í ÎåÄÎ¨∏ÏûêÎ°ú Î≥ÄÌôò
    return wrapper

@uppercase
def greet():
    return "hello, world!"

print(greet())
````
