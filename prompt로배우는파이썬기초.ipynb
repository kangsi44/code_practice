#### 2. 변수
변수는 값을 저장하는 데 사용되는 이름이며, 데이터를 참조하고 조작하는 데에 필요한 기본적인 요소입니다. 파이썬에서 변수를 사용하여 데이터를 저장하고, 이를 활용하여 계산하거나 조작할 수 있습니다. 변수는 다양한 데이터 타입을 가질 수 있으며, 값이 변경될 수 있는 특징을 가지고 있습니다.

변수를 사용하기 위해서는 변수를 선언하고 값을 할당해야 합니다. 파이썬에서는 변수 선언을 별도로 할 필요가 없으며, 변수에 값을 할당하는 과정에서 변수가 자동으로 생성됩니다. 변수 할당은 등호(=)를 사용하여 수행되며, 오른쪽의 값을 왼쪽의 변수에 저장합니다.

다음은 변수를 사용한 예제입니다.



```python
# 변수 할당
name = "Alice"
age = 25
pi = 3.14
is_student = True

# 변수 값 출력
print(name)         # "Alice" 출력
print(age)          # 25 출력
print(pi)           # 3.14 출력
print(is_student)   # True 출력

# 변수 값 변경
name = "Bob"
age = age + 1
pi = pi * 2
is_student = False

# 변경된 변수 값 출력
print(name)         # "Bob" 출력
print(age)          # 26 출력
print(pi)           # 6.28 출력
print(is_student)   # False 출력

```

위의 예제에서는 name, age, pi, is_student라는 변수를 선언하고 값을 할당했습니다. 각 변수에는 문자열, 정수, 부동 소수점 숫자, 불리언(Boolean) 값이 할당되었습니다. 변수 값은 print() 함수를 사용하여 출력되었습니다.

#### 4. 연산자
파이썬에서 연산자는 값을 조작하거나 연산을 수행하는 데 사용되는 기호입니다. 파이썬은 다양한 종류의 연산자를 제공하여 수치 연산, 논리 연산, 비교 연산 등을 수행할 수 있습니다.


1. 산술 연산자:
- `+` : 덧셈 연산을 수행합니다.
- `-` : 뺄셈 연산을 수행합니다.
- `*` : 곱셈 연산을 수행합니다.
- `/` : 나눗셈 연산을 수행하고 결과를 실수형으로 반환합니다.
- `//` : 나눗셈 연산을 수행하고 결과를 정수형으로 반환합니다.
- `%` : 나머지 연산을 수행합니다.
- `**` : 거듭제곱 연산을 수행합니다.

```python

a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.3333333333333335
print(a // b) # 3
print(a % b)  # 1
print(a ** b) # 1000
```


1. 할당 연산자:
- `=` : 변수에 값을 할당합니다.
- `+=` : 현재 변수에 값을 더하여 결과를 변수에 할당합니다.
- `-=` : 현재 변수에서 값을 빼고 결과를 변수에 할당합니다.
- `*=` : 현재 변수에 값을 곱하여 결과를 변수에 할당합니다.
- `/=` : 현재 변수를 나누고 결과를 변수에 할당합니다.
- `//=` : 현재 변수를 나눈 뒤 정수형으로 변환하여 결과를 변수에 할당합니다.
- `%=` : 현재 변수를 나눈 뒤 나머지를 결과로 변수에 할당합니다.

```python

x = 10

x += 5  # x = x + 5
print(x)  # 15

x -= 3  # x = x - 3
print(x)  # 12

x *= 2  # x = x * 2
print(x)  # 24

x /= 4  # x = x / 4
print(x)  # 6.0

x //= 2 # x = x // 2
print(x)  # 3

x %= 2  # x = x % 2
print(x)  # 1
```


1. 비교 연산자:
- `==` : 두 값이 같은지 비교합니다.
- `!=` : 두 값이 다른지 비교합니다.
- `>` : 왼쪽 값이 오른쪽 값보다 큰지 비교합니다.
- `<` : 왼쪽 값이 오른쪽 값보다 작은지 비교합니다.
- `>=` : 왼쪽 값이 오른쪽 값보다 크거나 같은지 비교합니다.
- `<=` : 왼쪽 값이 오른쪽 값보다 작거나 같은지 비교합니다.

```python

a = 5
b = 3

print(a == b)  # False
print(a != b)  # True
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False
```


1. 논리 연산자:
- `and` : 두 조건식이 모두 참일 때 참을 반환합니다.
- `or` : 두 조건식 중 하나라도 참이면 참을 반환합니다.
- `not` : 조건식의 결과를 반대로 반환합니다.

```python

x = 5
y = 3
z = 7

print(x > y and x < z)  # True
print(x > y or x > z)   # True
print(not(x > y))       # False
```



위의 예제에서는 주요한 연산자들을 다루었습니다. 파이썬은 다양한 연산자를 제공하므로 필요에 따라 추가적인 연산자를 학습하고 활용할 수 있습니다.

#### 5. 조건문

파이썬에서 조건문은 주어진 조건에 따라 코드 블록을 실행하거나 건너뛰는 데 사용됩니다. 주로 `if`, `elif`, `else` 키워드를 사용하여 작성합니다. 조건문은 프로그램이 특정 조건을 충족할 때 특정 동작을 수행하도록 하는데 유용합니다. 아래 예시를 통해 조건문을 자세히 살펴보겠습니다.



```python

age = 18

if age >= 20:
    print("성인입니다.")
elif age >= 13:
    print("청소년입니다.")
else:
    print("어린이입니다.")
```



위의 예제에서는 `age` 변수에 값을 할당하고, 해당 값에 따라 조건문이 실행됩니다.
- 첫 번째 조건 `age >= 20`은 참이 아니므로 건너뛰고 다음 조건으로 넘어갑니다.
- 두 번째 조건 `age >= 13`은 참이므로 해당 코드 블록이 실행되고 "청소년입니다."라는 메시지가 출력됩니다.
- 마지막으로 `else` 구문은 앞선 조건들이 모두 거짓일 경우 실행되며, 위 예제에서는 해당되지 않으므로 실행되지 않습니다.

조건문은 단일 `if` 문으로 시작할 수도 있고, 여러 개의 `elif` 문과 선택적인 `else` 문을 포함할 수도 있습니다. 이를 통해 여러 가지 조건을 처리하고 원하는 결과를 얻을 수 있습니다.

아래 예시는 조건문을 사용하여 성적에 따라 학점을 부여하는 예제입니다.

```python

score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print("학점:", grade)
```



위 예제에서는 `score` 변수에 값을 할당하고, 해당 값에 따라 조건문이 실행됩니다. `if`와 `elif` 조건들은 참인 경우 해당되는 학점을 `grade` 변수에 할당하고, `else` 조건은 모든 이전 조건들이 거짓일 경우 실행되어 'F'를 `grade` 변수에 할당합니다. 마지막으로, 학점이 출력됩니다.

이러한 방식으로 조건문을 사용하여 프로그램의 흐름을 제어하고, 주어진 조건에 따라 다른 동작을 수행할 수 있습니다.

````python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print("학점:", grade)
````

#### 7. 함수

튜플은 괄호 ( ) 를 사용하여 생성하며, 각 값들은 쉼표 ( , ) 로 구분합니다. 예를 들어, 다음과 같이 튜플을 생성할 수 있습니다:

```python

my_tuple = (1, 2, 3, 'apple', 'banana')
```



위의 예제에서 `my_tuple`은 정수와 문자열로 이루어진 튜플입니다. 튜플은 서로 다른 타입의 값들을 함께 저장할 수 있습니다.

튜플은 인덱싱과 슬라이싱을 통해 특정 값을 가져올 수 있습니다. 인덱스는 0부터 시작하며, 음수 인덱스를 사용하여 끝에서부터 접근할 수도 있습니다. 예를 들어:

```python

print(my_tuple[0])       # 1 출력
print(my_tuple[3])       # 'apple' 출력
print(my_tuple[-1])      # 'banana' 출력
print(my_tuple[1:4])     # (2, 3, 'apple') 출력
```



튜플은 값을 변경할 수 없기 때문에, 인덱싱을 통해 접근한 값은 수정할 수 없습니다. 하지만, 튜플을 이용한 패킹과 언패킹을 통해 값을 다룰 수 있습니다. 패킹은 여러 값을 하나의 튜플로 묶는 것을 의미하고, 언패킹은 튜플을 개별 변수로 분해하는 것을 의미합니다. 예를 들어:

```python

numbers = (1, 2, 3)
a, b, c = numbers     # 언패킹
print(a)              # 1 출력
print(b)              # 2 출력
print(c)              # 3 출력
```



위의 예제에서 `numbers` 튜플의 각 값을 변수 `a`, `b`, `c`에 언패킹하여 할당하였습니다.

튜플은 값의 변경이나 추가가 불가능하기 때문에, 데이터의 불변성을 유지해야할 때 유용합니다. 또한, 함수가 여러 개의 값을 반환할 때 튜플을 사용하여 반환할 수도 있습니다

````python
my_tuple = (1, 2, 3, 'apple', 'banana')
print(my_tuple[0])       # 1 출력
print(my_tuple[3])       # 'apple' 출력
print(my_tuple[-1])      # 'banana' 출력
print(my_tuple[1:4])     # (2, 3, 'apple') 출력

numbers = (1, 2, 3)
a, b, c = numbers     # 언패킹
print(a)              # 1 출력
print(b)              # 2 출력
print(c)              # 3 출력
````

empty_set = set()

파이썬에서 예외 처리는 try-except 블록을 사용하여 구현됩니다. try 블록 안에는 예외가 발생할 가능성이 있는 코드를 작성하고, except 블록은 예외가 발생했을 때 실행될 코드를 작성합니다. 예외가 발생하지 않으면 except 블록은 건너뛰고 프로그램의 실행이 계속됩니다.

```python

try:
    # 예외가 발생할 수 있는 코드

    # 예외가 발생했을 때 실행될 코드
```

예외 처리에서 가장 중요한 부분은 예외 종류입니다. 예외 종류는 예외의 유형을 지정하는데 사용되며, 특정 예외에 대한 처리 코드를 작성할 때 사용됩니다. 예를 들어, ZeroDivisionError 예외는 0으로 나누기 오류를 나타내며, FileNotFoundError 예외는 파일을 찾을 수 없는 오류를 나타냅니다.

```python

try:

    divisor = int(input("나누는 숫자를 입력하세요: "))
    result = dividend / divisor
    print("나눗셈 결과:", result)
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
```



위의 코드에서, 사용자가 0을 입력하면 "0으로 나눌 수 없습니다."라는 메시지가 출력됩니다. 이것은 ZeroDivisionError 예외를 처리하는 except 블록에서 실행되는 코드입니다.

이 예제는 단순한 형태의 예외 처리를 보여줍니다. 복잡한 프로그램에서는 다양한 예외 종류와 예외 처리 방법을 사용해야 할 수도 있습니다.

- 파이썬에서 제너레이터는 이터레이터(Iterator)를 생성하는 함수입니다. 제너레이터 함수는 일반적인 함수와 비슷하게 정의되지만, `return` 문 대신 `yield` 문을 사용하여 값을 반환합니다. `yield` 문은 함수의 실행을 일시 중단하고 값을 생성하며, 다시 호출되었을 때 중단된 지점부터 다시 실행됩니다.

간단한 제너레이터 함수 예제를 살펴보겠습니다. 아래의 예제는 1부터 시작하여 지정된 범위 내의 홀수를 반환하는 제너레이터 함수입니다.

```python

def generate_odd_numbers(n):
    i = 1
    while i <= n:
        yield i
        i += 2
```



위의 예제에서 `generate_odd_numbers` 함수는 `n`이라는 인자를 받습니다. 함수 내부에서는 `while` 루프를 사용하여 `i` 값을 1씩 증가시키면서 홀수를 생성합니다. `yield i` 문은 홀수를 생성하고 함수 실행을 일시 중단합니다.

제너레이터 함수를 호출하면 제너레이터 객체가 반환됩니다. 이 객체는 이터레이터이기 때문에 `next()` 함수를 사용하여 값을 하나씩 가져올 수 있습니다. 예를 들어, 다음과 같이 제너레이터 객체를 생성하고 값을 가져오는 방법을 보여드리겠습니다.

```python

numbers = generate_odd_numbers(10)

print(next(numbers))  # 1
print(next(numbers))  # 3
print(next(numbers))  # 5
```



`next()` 함수를 호출할 때마다 제너레이터 함수는 `yield` 문까지 실행되고, `yield` 문에서 반환한 값을 가져옵니다. 이런식으로 필요한 만큼의 값을 생성하고 사용할 수 있습니다.

- 데코레이터(Decorator)는 파이썬에서 함수나 클래스를 수정하거나 기능을 추가하기 위해 사용되는 기능입니다. 데코레이터는 기존의 코드를 변경하지 않고도 기능을 추가하거나 수정할 수 있어 코드의 재사용성과 가독성을 높일 수 있습니다. 데코레이터는 함수나 클래스 위에 `@데코레이터_함수`와 같은 형태로 사용되며, 해당 함수는 데코레이팅된 함수나 클래스를 인자로 받아 기능을 추가하거나 수정한 뒤 반환합니다.

다음은 더 간단한 예제로 `@uppercase`라는 데코레이터를 사용하여 함수의 반환값을 대문자로 변환하는 예제입니다:

```python

def uppercase(func):
    def wrapper():
        result = func()  # 원래 함수 호출
        return result.upper()  # 반환값 대문자로 변환
    return wrapper

@uppercase
def greet():
    return "hello, world!"

print(greet())
```



위 예제에서 `uppercase` 데코레이터는 `wrapper`라는 내부 함수를 정의하고, 이 함수에서 원래 함수인 `greet`를 호출하고 반환값을 대문자로 변환하여 반환합니다. `@uppercase` 데코레이터를 사용하여 `greet` 함수를 장식하면, `greet` 함수의 반환값이 대문자로 변환된 상태로 출력됩니다.

출력 결과:

```

HELLO, WORLD!
```



이처럼 데코레이터를 사용하면 함수나 클래스의 동작을 수정하거나 기능을 추가할 수 있습니다.


````python
def uppercase(func):
    def wrapper():
        result = func()  # 원래 함수 호출
        return result.upper()  # 반환값 대문자로 변환
    return wrapper

@uppercase
def greet():
    return "hello, world!"

print(greet())
````

````python
# bard에서 만든 프롬프트 코드
def timeit(func):
  def wrapper(*args, **kwargs):
    start = time.time()
    result = func(*args, **kwargs)
    end = time.time()
    print(f"{func.__name__} took {end - start} seconds to run.")
    return result
  return wrapper

@timeit
def sum_to_1000():
  """1부터 1000까지 더하는 함수."""
  sum = 0
  for i in range(1, 1001):
    sum += i
  return sum

if __name__ == "__main__":
  print(sum_to_1000())
````
